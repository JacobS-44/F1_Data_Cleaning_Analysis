```python
#Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load data from our csv file
data = pd.read_csv('data/raw/f1_drivers.csv')
```

```python
#Explore the data. I will be looking for the first few rows of the dataset to understand the data structure, do an initial data quality
#check, and identify any patterns or anomalies in the data. 

#Check the first few rows of data
print(data.head())

#Get a summary of the dataset 
print(data.info())
```

```python
#Check for missing values
missing_values = data.isnull().sum()
print(missing_values)

#Fill missing values
data.fillna(0, inplace=True)
```

```python
#Data visualizations
plt.figure(figsize=(10, 6))
sns.histplot(data['surname'])
plt.title('Distribution of Surname')
plt.show()
```
