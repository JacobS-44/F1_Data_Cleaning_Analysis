```python
#Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load data from our csv file
data = pd.read_csv('data/raw/f1_drivers.csv')
```

```python
#Explore the data. I will be looking for the first few rows of the dataset to understand the data structure, do an initial data quality
#check, and identify any patterns or anomalies in the data. 

#Check the first few rows of data
print(data.head())

#Get a summary of the dataset 
print(data.info())
```

```python
#Check for missing values
missing_values = data.isnull().sum()
print(missing_values)

#Fill missing values
data.fillna(0, inplace=True)
```

```python
#Data visualizations

nationality_counts = data['nationality'].value_counts().reset_index()
nationality_counts.columns = ['nationality', 'count']  # Rename columns for clarity

# Get the top N nationalities (e.g., top 10)
top_nationalities = nationality_counts.nlargest(10, 'count')

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(data=top_nationalities, x='nationality', y='count')  # Count of drivers by nationality
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.title('Top 10 nationalities of Drivers')
plt.xlabel('nationalities')
plt.ylabel('Number of Drivers')
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
```
